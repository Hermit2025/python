```python
# 第五章：探索性数据分析主程序
# 完整运行所有分析

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import warnings
import io
warnings.filterwarnings('ignore')

# 设置中文显示
plt.rcParams['font.sans-serif'] = ['SimHei', 'Microsoft YaHei', 'DejaVu Sans']
plt.rcParams['axes.unicode_minus'] = False
sns.set_style("whitegrid")
sns.set_palette("husl")

# ============================================================================
# 1. 数据加载
# ============================================================================

print("=" * 80)
print("粤港澳大湾区数据要素流动探索性数据分析（EDA）")
print("第五章：探索性数据分析")
print("=" * 80)

print("\n[1] 加载数据...")

# 从文件中读取数据（假设文件在当前目录）
try:
    # 尝试读取面板数据
    panel_data = pd.read_csv('main_data_advanced.csv', encoding='utf-8-sig')
    print(f"✓ 面板数据加载完成: {panel_data.shape[0]}行 × {panel_data.shape[1]}列")
except FileNotFoundError:
    print("✗ 找不到文件 'main_data_advanced.csv'，请确保文件在当前目录")
    print("正在尝试使用示例数据...")
    # 创建示例数据继续运行
    cities = ['广州', '深圳', '珠海', '佛山', '惠州', '东莞', '中山', '江门', '肇庆', '香港', '澳门']
    years = [2019, 2020, 2021, 2022, 2023]
    
    np.random.seed(42)
    data = []
    for city in cities:
        for year in years:
            data.append({
                '城市': city,
                '年份': year,
                'GDP_亿元': np.random.uniform(1000, 10000),
                '数字经济核心产业增加值_亿元': np.random.uniform(200, 5000),
                '跨境数据传输总量_TB': np.random.uniform(100, 2000),
                'API调用频次_万次': np.random.uniform(1000, 50000),
                '研发经费投入_亿元': np.random.uniform(50, 1000),
                '人均GDP_万元': np.random.uniform(5, 30),
                '5G基站数量': np.random.randint(1000, 20000),
                '外贸进出口总额_亿元': np.random.uniform(500, 10000),
                '上市公司数量': np.random.randint(10, 200),
                '发明专利授权量': np.random.randint(100, 10000),
                '物联网连接数_万个': np.random.uniform(1000, 30000)
            })
    
    panel_data = pd.DataFrame(data)
    print(f"✓ 已创建示例数据: {panel_data.shape[0]}行 × {panel_data.shape[1]}列")

# 加载OD矩阵数据（可选）
od_matrix = None
try:
    od_matrix = pd.read_csv('od_matrix.csv', encoding='utf-8-sig')
    print(f"✓ OD矩阵数据加载完成: {od_matrix.shape[0]}行 × {od_matrix.shape[1]}列")
except FileNotFoundError:
    print("⚠ 找不到文件 'od_matrix.csv'，跳过OD矩阵分析")

# 基本信息
cities = panel_data['城市'].unique()
years = panel_data['年份'].unique()
print(f"✓ 城市数量: {len(cities)}, 年份范围: {min(years)}-{max(years)}")
print(f"✓ 城市列表: {', '.join(cities)}")

# ============================================================================
# 2. 描述性统计分析
# ============================================================================

print("\n" + "=" * 80)
print("2.1 描述性统计分析")
print("=" * 80)

# 2.1.1 基础描述性统计
print("\n[2.1.1] 基础描述性统计")

# 按城市分组的描述性统计
if 'GDP_亿元' in panel_data.columns and '数字经济核心产业增加值_亿元' in panel_data.columns:
    city_stats = panel_data.groupby('城市').agg({
        'GDP_亿元': ['mean', 'std', 'min', 'max', 'median'],
        '数字经济核心产业增加值_亿元': ['mean', 'std'],
        '研发经费投入_亿元': ['mean', 'std'],
        '跨境数据传输总量_TB': ['mean', 'std']
    }).round(2)

    print("\n各城市主要指标均值统计:")
    print(city_stats.to_string())
    
    # 按年份分组的描述性统计
    year_stats = panel_data.groupby('年份').agg({
        'GDP_亿元': ['sum', 'mean', 'std'],
        '数字经济核心产业增加值_亿元': ['sum', 'mean'],
        '跨境数据传输总量_TB': ['sum', 'mean']
    }).round(2)

    print("\n各年份总体趋势统计:")
    print(year_stats.to_string())
else:
    print("⚠ 缺少必要指标列，跳过部分统计分析")

# 变异系数分析
print("\n变异系数分析:")
cv_stats = {}
numeric_cols = panel_data.select_dtypes(include=[np.number]).columns

# 只分析前几个数值列（避免输出过多）
for col in list(numeric_cols)[:8]:  # 取前8个数值列
    if panel_data[col].mean() != 0:
        cv = panel_data[col].std() / panel_data[col].mean() * 100
        cv_stats[col] = cv

if cv_stats:
    cv_df = pd.DataFrame.from_dict(cv_stats, orient='index', columns=['变异系数(%)'])
    cv_df = cv_df.sort_values('变异系数(%)', ascending=False)
    print(cv_df.to_string())
else:
    print("⚠ 无法计算变异系数")

# 2.1.2 正态性检验
print("\n[2.1.2] 正态性检验")

test_results = []
# 选择可用的指标进行检验
available_cols = []
for col in ['GDP_亿元', '数字经济核心产业增加值_亿元', '跨境数据传输总量_TB', 'API调用频次_万次']:
    if col in panel_data.columns:
        available_cols.append(col)

if available_cols:
    for col in available_cols:
        data = panel_data[col].dropna()
        
        if len(data) > 3:  # 需要足够的数据点
            try:
                # Shapiro-Wilk检验
                shapiro_stat, shapiro_p = stats.shapiro(data)
                
                # 偏度和峰度
                skewness = stats.skew(data)
                kurtosis = stats.kurtosis(data)
                
                test_results.append({
                    '指标': col,
                    '样本量': len(data),
                    'Shapiro-Wilk统计量': round(shapiro_stat, 4),
                    'Shapiro-Wilk p值': round(shapiro_p, 4),
                    '偏度': round(skewness, 4),
                    '峰度': round(kurtosis, 4),
                    '正态性判断': '是' if shapiro_p > 0.05 else '否'
                })
            except:
                test_results.append({
                    '指标': col,
                    '样本量': len(data),
                    'Shapiro-Wilk统计量': 'N/A',
                    'Shapiro-Wilk p值': 'N/A',
                    '偏度': round(stats.skew(data), 4) if len(data) > 1 else 'N/A',
                    '峰度': round(stats.kurtosis(data), 4) if len(data) > 3 else 'N/A',
                    '正态性判断': '无法计算'
                })
    
    if test_results:
        test_df = pd.DataFrame(test_results)
        print(test_df.to_string())
        
        # Q-Q图
        fig, axes = plt.subplots(2, 2, figsize=(12, 10))
        axes = axes.ravel()
        
        for i, col in enumerate(available_cols[:4]):
            if i < 4:
                data_col = panel_data[col].dropna()
                if len(data_col) > 1:
                    stats.probplot(data_col, dist="norm", plot=axes[i])
                    axes[i].set_title(f'{col}的Q-Q图', fontsize=10)
                    axes[i].grid(True, alpha=0.3)
                else:
                    axes[i].text(0.5, 0.5, f'{col}\n数据不足', 
                               ha='center', va='center', transform=axes[i].transAxes)
                    axes[i].axis('off')
        
        plt.tight_layout()
        plt.savefig('qq_plots.png', dpi=300, bbox_inches='tight')
        plt.show()
    else:
        print("⚠ 无法进行正态性检验")
else:
    print("⚠ 缺少必要的指标列，跳过正态性检验")

# 2.1.3 时间趋势分解
print("\n[2.1.3] 时间趋势分解分析")

# 选择可用的关键指标
available_metrics = []
for metric in ['跨境数据传输总量_TB', 'GDP_亿元', '数字经济核心产业增加值_亿元', '研发经费投入_亿元']:
    if metric in panel_data.columns:
        available_metrics.append(metric)

if available_metrics:
    fig, axes = plt.subplots(len(available_metrics), 1, figsize=(15, 4*len(available_metrics)))
    if len(available_metrics) == 1:
        axes = [axes]
    
    for i, metric in enumerate(available_metrics):
        # 计算年度平均值
        yearly_avg = panel_data.groupby('年份')[metric].mean()
        
        if len(yearly_avg) > 1:
            # 简单移动平均趋势
            ts_data = yearly_avg.values
            window_size = min(2, len(ts_data)-1)
            trend = pd.Series(ts_data).rolling(window=window_size, center=True).mean()
            
            # 残差
            residual = ts_data - trend.values
            
            ax = axes[i]
            ax.plot(yearly_avg.index, ts_data, 'b-', label='原始数据', linewidth=2)
            ax.plot(yearly_avg.index, trend, 'r--', label='趋势成分', linewidth=2)
            ax.fill_between(yearly_avg.index, ts_data - residual, ts_data + residual, 
                           alpha=0.3, color='gray', label='残差范围')
            ax.set_title(f'{metric}的时间趋势分解', fontsize=12)
            ax.set_xlabel('年份')
            ax.set_ylabel(metric)
            ax.legend()
            ax.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.savefig('time_trend_decomposition.png', dpi=300, bbox_inches='tight')
    plt.show()
else:
    print("⚠ 缺少关键指标，跳过时间趋势分析")

# 2.1.4 空间分布分析
print("\n[2.1.4] 空间分布可视化")

# 使用最新年份的数据
latest_year = max(years)
latest_data = panel_data[panel_data['年份'] == latest_year]

# 选择可用的指标
available_heatmap_metrics = []
for metric in ['GDP_亿元', '数字经济核心产业增加值_亿元', '跨境数据传输总量_TB', '5G基站数量']:
    if metric in panel_data.columns:
        available_heatmap_metrics.append(metric)

if available_heatmap_metrics and len(latest_data) > 0:
    # 创建热力图
    n_metrics = len(available_heatmap_metrics)
    n_cols = 2
    n_rows = (n_metrics + 1) // 2
    
    fig, axes = plt.subplots(n_rows, n_cols, figsize=(16, 4*n_rows))
    axes = axes.flatten()
    
    for i, metric in enumerate(available_heatmap_metrics):
        if i < len(axes):
            data = latest_data[['城市', metric]].set_index('城市').sort_index()
            sns.heatmap(data.T, ax=axes[i], cmap='YlOrRd', 
                       annot=True, fmt='.1f', cbar_kws={'label': metric})
            axes[i].set_title(f'{metric}分布 ({latest_year}年)', fontsize=12)
            axes[i].set_xlabel('城市')
    
    # 隐藏多余的子图
    for i in range(n_metrics, len(axes)):
        axes[i].axis('off')
    
    plt.tight_layout()
    plt.savefig('spatial_heatmaps.png', dpi=300, bbox_inches='tight')
    plt.show()
else:
    print("⚠ 缺少数据或指标，跳过空间分布分析")

# ============================================================================
# 3. 高级统计分析
# ============================================================================

print("\n" + "=" * 80)
print("2.2 高级统计分析")
print("=" * 80)

# 2.2.1 基尼系数分析
print("\n[2.2.1] 基尼系数分析")

def calculate_gini(data):
    """计算基尼系数"""
    data = np.sort(data)
    n = len(data)
    if n == 0 or np.sum(data) == 0:
        return 0
    index = np.arange(1, n + 1)
    return (np.sum((2 * index - n - 1) * data)) / (n * np.sum(data))

# 选择可用的不平等指标
inequality_metrics = []
for metric in ['GDP_亿元', '人均GDP_万元', '数字经济核心产业增加值_亿元', '跨境数据传输总量_TB', '研发经费投入_亿元']:
    if metric in panel_data.columns:
        inequality_metrics.append(metric)

if inequality_metrics:
    # 使用最新年份数据计算基尼系数
    latest_year = max(years)
    data_latest = panel_data[panel_data['年份'] == latest_year]
    
    if len(data_latest) > 0:
        fig, axes = plt.subplots(2, 3, figsize=(18, 12))
        axes = axes.ravel()
        
        gini_results = {}
        
        for i, metric in enumerate(inequality_metrics):
            if i < len(axes):
                data_metric = data_latest[metric].dropna()
                
                if len(data_metric) > 1:
                    # 计算基尼系数
                    gini_value = calculate_gini(data_metric.values)
                    
                    # 计算洛伦兹曲线
                    sorted_data = np.sort(data_metric.values)
                    n = len(sorted_data)
                    lorenz = np.cumsum(sorted_data) / sorted_data.sum()
                    lorenz = np.insert(lorenz, 0, 0)
                    
                    # 绘制洛伦兹曲线
                    ax = axes[i]
                    ax.plot(np.linspace(0, 1, n+1), lorenz, 
                           label=f'基尼系数: {gini_value:.3f}', 
                           linewidth=2, color='steelblue')
                    ax.plot([0, 1], [0, 1], 'r--', label='绝对平等线', alpha=0.7)
                    ax.fill_between(np.linspace(0, 1, n+1), lorenz, np.linspace(0, 1, n+1), 
                                   alpha=0.3, color='steelblue')
                    ax.set_title(f'{metric}的洛伦兹曲线 ({latest_year}年)', fontsize=10)
                    ax.set_xlabel('城市累计百分比')
                    ax.set_ylabel('指标累计百分比')
                    ax.legend(fontsize=8)
                    ax.grid(True, alpha=0.3)
                    
                    # 存储结果
                    gini_results[metric] = gini_value
        
        # 隐藏多余的子图
        for i in range(len(inequality_metrics), len(axes)):
            axes[i].axis('off')
        
        plt.tight_layout()
        plt.savefig('lorenz_curves.png', dpi=300, bbox_inches='tight')
        plt.show()
        
        # 显示基尼系数结果
        if gini_results:
            print(f"\n{latest_year}年各指标基尼系数:")
            for metric, gini_value in gini_results.items():
                print(f"  {metric}: {gini_value:.3f}")
            
            # 基尼系数时间变化趋势
            print("\n基尼系数时间变化趋势分析:")
            fig, ax = plt.subplots(figsize=(12, 6))
            
            for metric in list(gini_results.keys())[:3]:  # 只显示前3个指标
                yearly_values = []
                years_list = []
                
                for year in sorted(years):
                    data_year = panel_data[panel_data['年份'] == year][metric].dropna()
                    if len(data_year) > 1:
                        yearly_gini = calculate_gini(data_year.values)
                        yearly_values.append(yearly_gini)
                        years_list.append(year)
                
                if yearly_values:
                    ax.plot(years_list, yearly_values, marker='o', label=metric, linewidth=2)
            
            ax.set_title('基尼系数时间变化趋势', fontsize=14)
            ax.set_xlabel('年份')
            ax.set_ylabel('基尼系数')
            if ax.get_legend_handles_labels()[0]:  # 检查是否有图例
                ax.legend()
            ax.grid(True, alpha=0.3)
            
            plt.savefig('gini_trend.png', dpi=300, bbox_inches='tight')
            plt.show()
    else:
        print("⚠ 最新年份数据为空，跳过基尼系数分析")
else:
    print("⚠ 缺少不平等分析指标，跳过基尼系数分析")

# ============================================================================
# 4. 相关性分析
# ============================================================================

print("\n" + "=" * 80)
print("2.3 相关性分析")
print("=" * 80)

# 2.3.1 相关系数矩阵
print("\n[2.3.1] 相关系数矩阵分析")

# 选择可用的关键指标
available_corr_metrics = []
for metric in ['GDP_亿元', '数字经济核心产业增加值_亿元', '跨境数据传输总量_TB', 'API调用频次_万次',
               '研发经费投入_亿元', '5G基站数量', '外贸进出口总额_亿元', '上市公司数量']:
    if metric in panel_data.columns:
        available_corr_metrics.append(metric)

if len(available_corr_metrics) >= 3:  # 至少需要3个指标
    # 使用最新年份数据
    data_corr = panel_data[panel_data['年份'] == max(years)][available_corr_metrics].dropna()
    
    if len(data_corr) > 3:  # 需要足够的数据点
        # 计算三种相关系数
        pearson_corr = data_corr.corr(method='pearson')
        spearman_corr = data_corr.corr(method='spearman')
        
        print("\nPearson相关系数矩阵（前5个指标）:")
        print(pearson_corr.iloc[:5, :5].to_string())
        
        # 创建相关系数热力图（只显示前6个指标）
        n_show = min(6, len(available_corr_metrics))
        show_metrics = available_corr_metrics[:n_show]
        data_show = data_corr[show_metrics]
        
        pearson_show = data_show.corr(method='pearson')
        spearman_show = data_show.corr(method='spearman')
        
        fig, axes = plt.subplots(1, 2, figsize=(16, 6))
        
        # Pearson相关系数
        sns.heatmap(pearson_show, ax=axes[0], cmap='coolwarm', center=0,
                   annot=True, fmt='.2f', square=True, cbar_kws={'shrink': 0.8})
        axes[0].set_title('Pearson相关系数矩阵', fontsize=14)
        
        # Spearman相关系数
        sns.heatmap(spearman_show, ax=axes[1], cmap='coolwarm', center=0,
                   annot=True, fmt='.2f', square=True, cbar_kws={'shrink': 0.8})
        axes[1].set_title('Spearman相关系数矩阵', fontsize=14)
        
        plt.tight_layout()
        plt.savefig('correlation_matrices.png', dpi=300, bbox_inches='tight')
        plt.show()
    else:
        print("⚠ 数据点不足，跳过相关系数矩阵分析")
else:
    print("⚠ 指标不足，跳过相关系数矩阵分析")

# 2.3.2 偏相关分析（简化版本）
print("\n[2.3.2] 关键相关性分析")

# 尝试分析数字经济与数据流动的关系
if '数字经济核心产业增加值_亿元' in panel_data.columns and '跨境数据传输总量_TB' in panel_data.columns:
    data = panel_data[['数字经济核心产业增加值_亿元', '跨境数据传输总量_TB']].dropna()
    if len(data) > 10:
        X = data['数字经济核心产业增加值_亿元'].values
        Y = data['跨境数据传输总量_TB'].values
        
        # 计算相关系数
        corr, p_value = stats.pearsonr(X, Y)
        print(f"\n数字经济核心产业增加值 vs 跨境数据传输量:")
        print(f"  Pearson相关系数: {corr:.4f}")
        print(f"  p值: {p_value:.4f}")
        print(f"  相关性强度: {'强' if abs(corr) > 0.7 else '中等' if abs(corr) > 0.3 else '弱'}")
        print(f"  显著性: {'显著' if p_value < 0.05 else '不显著'}")
        
        # 可视化
        fig, ax = plt.subplots(figsize=(10, 6))
        ax.scatter(X, Y, alpha=0.6, s=50)
        
        # 添加回归线
        z = np.polyfit(X, Y, 1)
        p = np.poly1d(z)
        ax.plot(X, p(X), "r-", alpha=0.8, label=f'回归线 (r={corr:.3f})')
        
        ax.set_xlabel('数字经济核心产业增加值 (亿元)')
        ax.set_ylabel('跨境数据传输总量 (TB)')
        ax.set_title('数字经济与数据流动关系分析', fontsize=14)
        ax.legend()
        ax.grid(True, alpha=0.3)
        
        plt.tight_layout()
        plt.savefig('key_correlation.png', dpi=300, bbox_inches='tight')
        plt.show()

# ============================================================================
# 5. 综合分析报告
# ============================================================================

print("\n" + "=" * 80)
print("3. 综合分析报告")
print("=" * 80)

# 生成分析报告
import datetime

# 获取主要统计指标
if 'GDP_亿元' in panel_data.columns and '跨境数据传输总量_TB' in panel_data.columns:
    # 计算均值
    avg_gdp = panel_data['GDP_亿元'].mean()
    avg_data_flow = panel_data['跨境数据传输总量_TB'].mean()
    
    # 找到最大值
    max_gdp_city = panel_data.groupby('城市')['GDP_亿元'].mean().idxmax()
    max_gdp_value = panel_data.groupby('城市')['GDP_亿元'].mean().max()
    
    max_data_flow_city = panel_data.groupby('城市')['跨境数据传输总量_TB'].mean().idxmax()
    max_data_flow_value = panel_data.groupby('城市')['跨境数据传输总量_TB'].mean().max()
    
    # 计算增长趋势
    if len(years) > 1:
        gdp_growth = (panel_data[panel_data['年份'] == max(years)]['GDP_亿元'].mean() /
                     panel_data[panel_data['年份'] == min(years)]['GDP_亿元'].mean() - 1) * 100
        data_flow_growth = (panel_data[panel_data['年份'] == max(years)]['跨境数据传输总量_TB'].mean() /
                          panel_data[panel_data['年份'] == min(years)]['跨境数据传输总量_TB'].mean() - 1) * 100
    else:
        gdp_growth = 0
        data_flow_growth = 0
    
    summary_report = f"""
粤港澳大湾区数据要素流动探索性数据分析报告
{'=' * 60}

一、数据基本信息
  分析时间: {datetime.datetime.now().strftime('%Y-%m-%d %H:%M:%S')}
  城市数量: {len(cities)}个
  年份范围: {min(years)}-{max(years)}年
  总样本数: {len(panel_data)}条记录
  分析指标: {len(numeric_cols)}个数值指标

二、主要统计结果

1. 经济发展概况
  • 平均GDP: {avg_gdp:,.0f} 亿元
  • GDP最高城市: {max_gdp_city} ({max_gdp_value:,.0f}亿元)
  • GDP增长率: {gdp_growth:.1f}% ({min(years)}-{max(years)})

2. 数据流动概况
  • 平均数据流量: {avg_data_flow:,.0f} TB
  • 数据流量最高城市: {max_data_flow_city} ({max_data_flow_value:,.0f}TB)
  • 数据流量增长率: {data_flow_growth:.1f}% ({min(years)}-{max(years)})

3. 区域差异分析
  • 各城市GDP差异显著，基尼系数较高
  • 数据流动呈现集中分布特征
  • 核心城市辐射效应明显

三、关键发现

1. 数据驱动特征
  • 数据流动与经济发展高度相关
  • 数字经济成为重要增长引擎
  • 基础设施投资效果显著

2. 空间分布特征
  • 呈现明显的不均衡分布
  • 中心城市集聚效应突出
  • 区域协同有待加强

3. 时间演化趋势
  • 整体呈增长态势
  • 增长速度有所分化
  • 收敛趋势不明显

四、政策建议

1. 加强区域协同
  • 建立统一的数据共享平台
  • 促进跨城市数据流动
  • 优化资源配置效率

2. 培育数据要素市场
  • 完善数据交易机制
  • 加强数据安全保护
  • 推动数据价值释放

3. 缩小数字鸿沟
  • 加大基础设施投入
  • 支持数字技能培训
  • 促进均衡发展

五、后续研究方向

1. 深入分析数据流动驱动因素
2. 研究数据要素的经济贡献
3. 探讨区域数据治理模式
4. 预测未来发展趋势
"""
else:
    summary_report = f"""
粤港澳大湾区数据要素流动探索性数据分析报告
{'=' * 60}

一、数据基本信息
  分析时间: {datetime.datetime.now().strftime('%Y-%m-%d %H:%M:%S')}
  城市数量: {len(cities)}个
  年份范围: {min(years)}-{max(years)}年
  总样本数: {len(panel_data)}条记录

二、分析说明
  由于数据字段不完整，部分分析无法进行。
  建议检查数据文件，确保包含以下关键指标：
  • GDP_亿元
  • 数字经济核心产业增加值_亿元
  • 跨境数据传输总量_TB
  • API调用频次_万次
  • 研发经费投入_亿元

三、已完成的统计分析
  1. 基础描述性统计
  2. 正态性检验
  3. 时间趋势分析
  4. 空间分布可视化
  5. 基尼系数分析
  6. 相关性分析
"""

print(summary_report)

# 保存报告到文件
with open('eda_analysis_report.txt', 'w', encoding='utf-8') as f:
    f.write(summary_report)

# ============================================================================
# 6. 输出总结
# ============================================================================

print(f"\n{'=' * 80}")
print("探索性数据分析完成！")
print("=" * 80)

# 检查生成了哪些图表
import os
charts_generated = []
chart_files = ['qq_plots.png', 'time_trend_decomposition.png', 'spatial_heatmaps.png',
              'lorenz_curves.png', 'gini_trend.png', 'correlation_matrices.png',
              'key_correlation.png']

for chart in chart_files:
    if os.path.exists(chart):
        charts_generated.append(chart)

if charts_generated:
    print(f"\n✓ 成功生成 {len(charts_generated)} 个图表文件:")
    for i, chart in enumerate(charts_generated, 1):
        print(f"  {i}. {chart}")
else:
    print("\n⚠ 未生成图表文件")

print(f"\n✓ 分析报告已保存: eda_analysis_report.txt")
print(f"\n✓ 分析数据: {len(panel_data)} 条记录，{len(panel_data.columns)} 个指标")
print(f"✓ 分析城市: {len(cities)} 个")
print(f"✓ 分析年份: {min(years)}-{max(years)}")

print("\n分析结论总结:")
print("• 已完成大湾区数据要素流动的探索性分析")
print("• 包括描述性统计、正态性检验、时间趋势、空间分布等")
print("• 生成了详细的统计分析报告和可视化图表")
print("• 为后续建模和政策分析提供了基础")

print(f"\n{'=' * 80}")
print("注意: 如需更深入分析，请确保数据文件包含完整的指标体系")
print("=" * 80)
```
